```python
#Imports
import tweepy
import json
import datetime as dt
```

```python
#Get Bearer Token
with open('config.json', 'r') as file:
    data = json.load(file)
    bearer_token = data['bearer_token']
```

```python
#Authenticate with twitter
auth = tweepy.OAuth2BearerHandler(bearer_token)
api = tweepy.API(auth)
```

```python
#Get Data
query = ["msft", "microsoft"]
results = api.search_tweets(q=query)
```

```python
#Save the data
queryName = str(query) + "__" + str(dt.datetime.now())
queryName = queryName.replace(' ', '_')
queryName = queryName.replace(':', '_')
queryName = queryName.replace('-', '_')

textData = [item.text for item in results]
print(textData)

with open("twitter_data/" + queryName + ".txt",'w', encoding='utf-8') as file:
    file.writelines(textData)
```
